```python
content['Category'].unique()
```

```python
content['Category'] = content['Category'].str.strip('"')
```

```python
content['Category'] = content['Category'].str.lower()
```

```python
content['Category'].unique()
```

```python
merged_df = pd.merge(reactions, content, left_on='Content ID', right_on= 'Content ID', how='left')
merged_df
```

```python
merged_df.rename(columns = {'Type_x':'ReactionType','Type_y':'ContentType'}, inplace= True)
merged_df
```

```python
merged_df_final = pd.merge(merged_df, reaction_types, left_on='ReactionType', right_on='Type', how ='left').drop('Type', axis=1)
merged_df_final
```

```python
merged_df_final.dtypes
```

```python
grouped_data = merged_df_final.groupby('Category')['Score'].sum().reset_index().sort_values(by='Score', ascending=False)
grouped_data.rename(columns = {'Category': 'Category_','Score':'Score_'}, inplace= True)
grouped_data
#
```

```python
result = pd.concat([merged_df_final, grouped_data], axis=1).sort_values(by='Score_', ascending=False)
result
```

```python
result.to_csv('end_result.csv')
```
